```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Dell\Desktop\Panda\2019.csv")
```

```python
print(data.isnull().sum())
```

```python
column_means = data.drop(columns='Country or region').mean()
```

```python
for column in data.columns:
    if column != 'Country or region':
        # Replace 0 values with column mean
        data[column] = data[column].replace(0, column_means[column])
```

```python
data.to_csv(r"C:\Users\Dell\Desktop\Panda\2019.csv")
```

```python
top_countries=data.groupby("Country or region")["Overall rank"].sum().nsmallest(5)
plt.figure(figsize=(10,6))
top_countries.plot(kind="bar",rot=45,color="red")
plt.xlabel("Country or region")
plt.ylabel("rank")
plt.title("Rank of countries wrt Highest Happines Score")
plt.show()
top_countries=data.groupby("Country or region")["Overall rank"].sum().nlargest(5)
plt.figure(figsize=(10,6))
top_countries.plot(kind="bar",rot=45,color="red")
plt.xlabel("Country or region")
plt.ylabel("rank")
plt.title("Rank of countries wrt lowest Happines Score")
plt.show()
```
